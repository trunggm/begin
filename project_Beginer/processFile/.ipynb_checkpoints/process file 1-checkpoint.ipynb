```python
f = open('link.txt', 'r')
data = f.read()
f.close()
```

```python
len(data)
```

```python
arraylink = data.split('\n')
len(arraylink)
```

```python
arraylink = [x for x in arraylink if x.startswith('http://truyenfull.vn/')]
len(arraylink)
```

```python
arraylink.sort()
arraylink[0:100]
```

```python
arraylink = [x for x in arraylink if x.find('#')==-1]
len(arraylink)
```

```python
arraylink[0:100]
```

```python
arraylink = [x for x in arraylink if len(x.split('/'))==6]
arraylink[:17]
```

```python
f = open('linkchap.txt', 'a')
for link in arraylink:
    f.write(link+"\n")
f.close()
```
